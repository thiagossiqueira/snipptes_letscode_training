```python
#Exemplos
# Bibliotecas
from bqplot import pyplot as plt
import bql
import bqviz as bqv
import pandas as pd

# Instanciando objeto
bq = bql.Service()

# Definindo parâmetros e data items
params = {'dates':'range(2016-01-01, 2016-01-15)', 'fill':'next'}
px_open = bq.data.px_open(**params)
px_last = bq.data.px_last(**params)
px_high = bq.data.px_high(**params)
px_low = bq.data.px_low(**params)
intervalo = px_high - px_low

# Dicionário para os indices do dataframe
items_dic = {'open':px_open, 'last':px_last, 'high':px_high, 'low':px_low}

# Gerando e executando request
request = bql.Request('IBM US Equity', items_dic)
resposta = bq.execute(request)

# Transformando em um dataframe
df = bql.combined_df(resposta)
df.set_index('DATE', inplace=True)
df.head()

# Modificando a ordem das colunas do dataframe
df = pd.DataFrame(df, columns=['open', 'last', 'high', 'low', 'CURRENCY'])
df.head()

# Removendo coluna de cambios
df.drop('CURRENCY', inplace=True)

# Plot básico de linhas
figura = bqv.LinePlot(df).set_style()
figura.show()

# Barras
figura = bqv.BarPlot(df).set_style()
figura.show()

# Kernel Density Estimation
figura = bqv.KDEPlot(df, bandwidth=0.09).set_style()
figura.show()

# Histograma
figura = bqv.HistPlot(df, bins=20)
figura.show()

# Distribuição com a coluna 'low'
low = pd.DataFrame(df['low'])
figura = bqv.DistributionPlot(low).set_style()
figura.show()

# Novo request e dataframe
request = bql.Request('NFLX US Equity', items_dic)
resposta = bq.execute(request)
df2 = bql.combined_df(resposta)
df2.set_index('DATE', inplace=True)
df2.drop('CURRENCY', axis=1, inplace=True)
df2.head()

# Ambos são passados para gráfico de comparação
fig = bqv.MultiComparisonPlot(df, df2)
fig.show()

# Grid com vários tipos de plot
grid = [bqv.LinePlot, bqv.BarPlot, bqv.HistPlot, bqv.KDEPlot]
fig = bqv.GridPlot(df, plots=grid)
fig.show()

```
