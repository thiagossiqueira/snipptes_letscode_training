```python
#Exemplos
#Bibliotecas
import bql
import bqviz as bqv
from ipywidgets import *

bq = bql.Service()

# Universo
entidades = ['MSFT US Equity', 'GE US Equity', 'BAC US Equity']

# Data item
intervalo = bq.func.range('2019-01-01', '2019-12-31')
preco = bq.data.px_last(dates=intervalo, frq='D', fill='prev')

# Request
request = bql.Request(entidades, preco)
resposta = bq.execute(request)

# Dataframe
df = bql.combined_df(resposta)
df.set_index('DATE', inplace=True)
df.drop('CURRENCY', axis=1, inplace=True)


### Link básico
# Criando plot interativa
intlinhas = bqv.InteractiveLinePlot(df, title='Principal', hide_controls=True)

# Criando plot normal que irá recuperar os dados da anterior
linhas = bqv.LinePlot(title='Subplot')

# Linkando ambas
intlinhas.links = [linhas]

# Mostrando as duas em um box no formato vertical
VBox([intlinhas.show(), linhas.show()])

### Linkando a estatísticas
# Definindo função que retorna um df com as estatisticas
def estatisticas(df):
    dic = {'Min':df.min(), 'Max':df.max(), 'Media':df.mean()}
    df2 = pd.DataFrame(dic)
    df2.dropna(axis=1, inplace=True)
    return df2

# Criando barplot vazio
barras = bqv.BarPlot()

# O barplot vai receber as estatísticas referentes ao df no gráfico interativo de linhas
intlinhas.links = [[barras, estatisticas]]

# Mostrando o gráfico interativo e o gráfico de barras com as estatisticas
VBox([intlinhas.show(), barras.show()])


### Selected data

# Modificando dataframe para a dispersao
df.reset_index(inplace=True)

# Mais uma plot interativa
intdispersao = bqv.InteractiveScatterPlot(df, hide_controls=True)
intdispersao.show()

# Seleciona-se os dados e depois roda-se a seguinte célula:
intdispersao.selected_data
```
